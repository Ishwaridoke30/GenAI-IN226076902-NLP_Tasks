Task 2:Build Advanced Preprocessing Function
Implement the following function:
def preprocess_text(text):
    pass

Required Features
Your function must include:
Remove numbers
Example: "I have 2 dogs" → "I have dogs"
Remove extra spaces
Example: "This is good" → "This is good"
Handle repeated characters
Example: "soooo goooood!!!" → "so good"
Convert text to lowercase
Example: "WOW!!! This is GREAT!!!" → "wow this is great"
Remove very short tokens
Remove words with length ≤ 2
Exception: keep meaningful words like "no", "not"
Remove URLs and email-like patterns
Example: "Visit http://example.com now" → "visit now"



```python
import re
def preprocess_text(text):  
    
    if not text:
        return [], ""
    
    # 1. Convert to lowercase
    text = text.lower()    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)    
    # 3. Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)    
    # 5. Handle repeated characters (soooo → so, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Remove special characters / emojis
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()    
    # 8. Tokenize
    tokens = text.split()    
    # 9. Remove very short tokens (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]    
    # 10. Join back to sentence
    clean_sentence = " ".join(tokens)    
    return tokens, clean_sentence
```

Task 4: Token Analytics
For each processed sentence, compute:
Total number of tokens
Number of unique tokens
Average token length
Analysis Questions
Which sentence had the most noise?
Which sentence retained the most meaningful tokens after cleaning?


```python
def analyze_tokens(tokens):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0
    
    return total_tokens, unique_tokens, avg_length


# Run analysis for each sentence
for i, text in enumerate(sentences, 1):
    tokens, clean_sentence = preprocess_text(text)
    
    total, unique, avg = analyze_tokens(tokens)
    
    print(f"Sentence {i}")
    print("Original:", text)
    print("Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print("-" * 60)
```

```python
from collections import Counter
all_tokens = []
for text in sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)
# Countfreq
counter = Counter(all_tokens)
# Top 10 most frequent words
top_10 = counter.most_common(10)
# Top 5 least frequent words
least_5 = counter.most_common()[-5:]
print("Top 10 Most Frequent Words:")
print(top_10)
print("\nTop 5 Least Frequent Words:")
print(least_5)
```

Task 6: Build Full Pipeline
def full_pipeline(text_list):
    pass
Expected Output
{
    "tokens": [...],
    "clean_sentences": [...]
}



```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)        
        all_tokens.extend(tokens) # collect all tokens
        clean_sentences.append(clean_sentence)  # store cleaned sentence
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for s in result["clean_sentences"]:
    print(s)
    
```

Task 7: Error Handling
Ensure your solution handles:

Only emojis
Only numbers




```python
import re

def preprocess_text(text):
    
    # 1. Handle empty string
    if not text or text.strip() == "":
        return [], ""    
    # 2. Lowercase
    text = text.lower()    
    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)    
    # 4. Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    # 5. Remove numbers
    text = re.sub(r'\d+', '', text)    
    # 6. Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)    
    # 7. Remove special characters / emojis
    text = re.sub(r'[^a-z\s]', '', text)    
    # 8. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()    
    # 9. Tokenize
    tokens = text.split()    
    # 10. Remove short tokens except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]    
    # 11. If nothing left after cleaning
    if not tokens:
        return [], ""    
    clean_sentence = " ".join(tokens)    
    return tokens, clean_sentence
```

```python
test_cases = [
    "",                # empty
    "😍😍😂😂🔥🔥",      # only emojis
    "1234567890",      # only numbers
]

for i, text in enumerate(test_cases, 1):
    tokens, clean = preprocess_text(text)
    
    print(f"Test Case {i}")
    print("Original:", text)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print("-" * 40)
```
