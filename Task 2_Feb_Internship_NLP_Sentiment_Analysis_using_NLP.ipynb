```python
data = [
    ("I love this product", "positive"),
    ("This is amazing", "positive"),
    ("Worst service ever", "negative"),
    ("I am not happy", "negative"),
    ("Very good quality", "positive"),
    ("Bad experience", "negative"),
    ("Not good at all", "negative"),
    ("Absolutely fantastic", "positive")
]
```

```python
import re

stop_words = {"is","am","are","the","a","an","and","to","of","in","this","that"}

def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return words
```

```python
processed_data = []

for text, label in data:
    tokens = preprocess(text)
    processed_data.append((tokens, label))

print(processed_data)
```

```python
vocab = set()

for tokens, _ in processed_data:
    vocab.update(tokens)

vocab = list(vocab)
print("Vocabulary:", vocab)
```

```python
positive_words = {"good","love","amazing","fantastic","best"}
negative_words = {"bad","worst","not","poor"}

def predict(tokens):
    score = 0
    
    for word in tokens:
        if word in positive_words:
            score += 1
        elif word in negative_words:
            score -= 1
    
    return "positive" if score >= 0 else "negative"
```

```python
for tokens, actual in processed_data:
    pred = predict(tokens)
    print("Tokens:", tokens)
    print("Actual:", actual, "| Predicted:", pred)
    print()
```

```python
correct = 0

for tokens, actual in processed_data:
    if predict(tokens) == actual:
        correct += 1

accuracy = correct / len(processed_data)
print("Accuracy:", accuracy)
```

```python
print("- Preprocessing reduced noise in text")
print("- Rule-based model used for classification")
print("- Accuracy achieved on sample dataset")
print("- Model is simple but effective for small data")
```
