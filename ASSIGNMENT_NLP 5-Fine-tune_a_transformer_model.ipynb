```python
true_labels = [
    ["B-PER", "O", "O", "B-ORG", "O"],
    ["B-LOC", "O", "O"]
]

pred_labels = [
    ["B-PER", "O", "O", "B-ORG", "O"],
    ["B-LOC", "O", "O"]
]
```

```python
true_flat = [label for sent in true_labels for label in sent]
pred_flat = [label for sent in pred_labels for label in sent]
```

```python
from collections import Counter

def calculate_metrics(true, pred):
    correct = sum(t == p for t, p in zip(true, pred))
    total_pred = len(pred)
    total_true = len(true)

    precision = correct / total_pred if total_pred > 0 else 0
    recall = correct / total_true if total_true > 0 else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1

precision, recall, f1 = calculate_metrics(true_flat, pred_flat)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
def calculate_metrics_ignore_O(true, pred):
    filtered = [(t, p) for t, p in zip(true, pred) if t != "O"]

    if not filtered:
        return 0, 0, 0

    true_f, pred_f = zip(*filtered)

    correct = sum(t == p for t, p in zip(true_f, pred_f))
    precision = correct / len(pred_f)
    recall = correct / len(true_f)
    f1 = (2 * precision * recall) / (precision + recall)

    return precision, recall, f1

precision, recall, f1 = calculate_metrics_ignore_O(true_flat, pred_flat)

print("Precision (no O):", precision)
print("Recall (no O):", recall)
print("F1 Score (no O):", f1)
```

### Difference between POS Tagging and Chunking

Part-of-Speech (POS) Tagging assigns grammatical labels to each word in a sentence such as noun, verb, adjective, etc. It works at the word level and helps in understanding the basic grammar of a sentence.

Chunking, on the other hand, groups words into meaningful phrases such as noun phrases (NP) and verb phrases (VP). It works at the phrase level and provides deeper understanding of sentence structure.

In simple terms, POS tagging focuses on individual words, while chunking focuses on groups of words.
